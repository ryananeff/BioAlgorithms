```python
from collections import defaultdict

input_str="""ACGTTGCATGTCGCATGATGCATGAGAGCT
4""".strip().split("\n")

string1 = None
int1 = None
for line in input_str:
    if string1 == None:
        string1 = line
        continue
    if int1 == None:
        int1 = int(line)
        continue

kmer_dict = defaultdict(int)

for i in range(0,len(string1)-int1+1):
    kmer = string1[i:i+int1]
    kmer_dict[kmer] += 1

max_kmer_occur = max(kmer_dict.values())
sorted_kmer_occur = sorted(kmer_dict.items(), key=lambda x:x[1], reverse=True)

for a,b in sorted_kmer_occur:
    if b==max_kmer_occur:
        print(a, end=" ")
    else:
        break
print("\n")
```

```python
input_str="""GAATGCAACCACTTCGTTTAGTATTGAGCAGCTGAGCAGCACCACTTCACCACTTCGTTTAGTATATAGAAGCAACCACTTCATAGAAGCAATAGAAGCAACCACTTCTGAGCAGCATAGAAGCAACCACTTCACCACTTCTGAGCAGCTGAGCAGCATAGAAGCAGAATGCAGTTTAGTATACCACTTCGAATGCATGAGCAGCACCACTTCGTTTAGTATGTTTAGTATGTTTAGTATACCACTTCGAATGCAATAGAAGCAACCACTTCATAGAAGCAATAGAAGCAGAATGCAGTTTAGTATACCACTTCTGAGCAGCGAATGCAACCACTTCACCACTTCGAATGCAGAATGCAATAGAAGCAGTTTAGTATTGAGCAGCGTTTAGTATTGAGCAGCACCACTTCGAATGCAACCACTTCACCACTTCATAGAAGCAGTTTAGTATACCACTTCGAATGCAATAGAAGCAGTTTAGTATGTTTAGTATATAGAAGCAACCACTTCGAATGCAGTTTAGTATGTTTAGTATTGAGCAGCATAGAAGCAACCACTTCATAGAAGCAGTTTAGTATACCACTTCGTTTAGTATATAGAAGCAGTTTAGTATATAGAAGCAATAGAAGCAATAGAAGCAACCACTTCGTTTAGTATACCACTTCGTTTAGTATTGAGCAGCGAATGCAACCACTTCGAATGCAGAATGCAATAGAAGCAGAATGCAATAGAAGCATGAGCAGCACCACTTCGTTTAGTATGAATGCATGAGCAGCGAATGCATGAGCAGCATAGAAGCAGAATGCAGAATGCA
7 2""".strip().split("\n")

nucleotides = ["A","C","T","G"]

def hamming_distance(string1, string2):
    return sum([a!=b for a,b in zip(string1, string2)])

def generate_mismatches(string, nucleotides=nucleotides, number=1):
    newstr = [a for a in string]
    new_strings = [string]
    if number <= 0:
        return new_strings
    for s in range(len(string)):
        for n in range(len(nucleotides)):
            if string[s]!=nucleotides[n]:
                newstr[s] = nucleotides[n]
                gen_string = "".join(newstr)
                if number > 1:
                    strings = generate_mismatches(gen_string, number=number-1)
                    for ss in range(len(strings)):
                        sstr = [a for a in strings[ss]]
                        sstr[s] = nucleotides[n]
                        strings[ss] = "".join(sstr)
                    strings = list(set(strings))
                    for ss in strings:
                        new_strings.append(ss)
                else:
                    new_strings.append(gen_string)
        newstr = [a for a in string]
    return new_strings

kmer_dict = defaultdict(int)
string1 = None
nmer = None
mismatch = None
for line in input_str:
    if string1 == None:
        string1 = line
        continue
    else:
        nmer, mismatch = [int(i) for i in line.split(" ")]

for i in range(len(string1)-nmer+1):
    kmer = string1[i:i+nmer]
    kmer_dict[kmer] += 1

sorted_kmer_occur = sorted(kmer_dict.items(), key=lambda x:x[1], reverse=True)

kmer_dict_max = defaultdict(int)
for t,occ in sorted_kmer_occur:
    new_kmers = generate_mismatches(t,number=mismatch)
    for n in new_kmers:
        kmer_dict_max[n] += occ

max_kmer_occur_max = max(kmer_dict_max.values())
sorted_kmer_occur_max = sorted(kmer_dict_max.items(), key=lambda x:x[1], reverse=True)

for a,b in sorted_kmer_occur_max:
    if b==max_kmer_occur_max:
        print(a, end=" ")
    else:
        break
print("\n")
```
